```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
from tensorflow.keras.preprocessing import image
import os
import matplotlib.pyplot as plt
import cv2
import warnings
import kagglehub
import requests
warnings.filterwarnings('ignore')
```

```python
# Declare variables
image_dims = 128
batch_size = 64
```

```python
# Create model
model = Sequential()

model.add(Conv2D(64, (3,3), activation = 'relu', input_shape = (image_dims, image_dims, 3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(128, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
input_path = 'chest_xray'
```

```python
# Training data
training_data_generator =  ImageDataGenerator(rescale = 1./255,
                                             shear_range = 0.2,
                                             zoom_range = 0.2,
                                             horizontal_flip= True)

training_gen = training_data_generator.flow_from_directory(directory=input_path+'\\train',
                             target_size=(image_dims,image_dims),
                             batch_size=batch_size,
                             class_mode='binary')
```

```python
# Test data:

validation_data_generator = ImageDataGenerator(rescale= 1./255)

validation_gen = validation_data_generator.flow_from_directory(directory= input_path+ '\\val',
                                                     target_size=(image_dims,image_dims),
                                                     batch_size= batch_size,
                                                     class_mode= 'binary')
```

```python
epochs = 10
history = model.fit(training_gen,
             steps_per_epoch= 10,
             epochs = epochs,
             validation_data=validation_gen,
             validation_steps= validation_gen.samples)
```

```python
plt.figure(figsize=(8,6))
plt.title('Accuracy scores')
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title('Loss value')
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()
```

```python
test_data_generator = ImageDataGenerator(rescale = 1/255)

test_gen = test_data_generator.flow_from_directory(directory = input_path + '\\test',
    target_size = (image_dims, image_dims),
    batch_size = 128, 
    class_mode = 'binary'
)
```

```python
eval_result = model.evaluate(test_gen, steps = 624)
print('Loss :', eval_result[0])
print('Accuracy :', eval_result[1])
```

Our deep learning model has achieved an accuracy of 85% in classifying images. This indicates that the model performs well in recognizing patterns in unseen data. We can leverage this model to predict whether an individual is healthy or has pneumonia based on their chest X-ray images.
