### Preprocessing

```python
client_prod = train[train['Semana'] <= 8].groupby(['Producto_ID', 'Cliente_ID']).agg({'Demanda_uni_equil': ['mean', 'count']
}).reset_index()
```

```python
client_prod.columns = ['Producto_ID', 'Cliente_ID', 'mean_client_prod', 'count_client_prod']
```

```python
client_prod.head()
```

```python
df = df.merge(client_prod, on = ['Producto_ID', 'Cliente_ID'], how = 'left')
```

```python
prod = train[train['Semana'] <= 8].groupby(['Producto_ID']).agg({'Demanda_uni_equil': ['mean', 'count']}).reset_index()
```

```python
prod.columns = ['Producto_ID', 'mean_prod', 'count_prod']
```

```python
df = df.merge(prod, on = ['Producto_ID'], how = 'left')
```

```python
client = train[train['Semana'] <= 8].groupby(['Cliente_ID']).agg({'Demanda_uni_equil': ['mean', 'count']}).reset_index()
```

```python
client.columns = ['Cliente_ID', 'mean_cliente', 'count_cliente']
```

```python
df = df.merge(client, on = ['Cliente_ID'],how = 'left')
```

### Modeling

```python
feature_cols = ['Semana', 'Producto_ID', 'Cliente_ID', 'mean_client_prod', 'count_client_prod',
                'mean_prod', 'count_prod', 'mean_cliente', 'count_cliente']
```

```python
train_mask = df['tst'] == 0
```

```python
y_train = df.loc[train_mask, 'Demanda_uni_equil']
```

```python
X_train = df.loc[train_mask, feature_cols]
```

```python
X_test = df.loc[~train_mask, feature_cols]
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
params = {
    'max_depth': 8,                  # Maximum depth of a tree
    'eta': 0.5,                      # Learning rate
    'objective': 'reg:squarederror'  # Objective function for regression
}
```

```python
model = xgb.train(params, dtrain, num_boost_round=25)
```

```python
dtest = xgb.DMatrix(X_test)
```

```python
preds = model.predict(dtest)
```

```python
preds = np.maximum(preds, 0)
```

```python
submission = pd.DataFrame({
    'id': df.loc[~train_mask, 'id'].astype(int),
    'Demanda_uni_equil': preds
})
```
